```python
def find_row_pairs(df, num_diff):
    pairs = []

    # iterate over each pair of rows in the dataframe
    for i, row1 in df.iterrows():
        for j, row2 in df.iterrows():
            if i < j:  # avoid checking same row pair twice
                diff = (row1 != row2)  # boolean array of differences between rows
                if diff.sum() == num_diff:
                    cols = diff[diff].index.tolist()  # get column names where the rows differ
                    pairs.append((i, j, cols))
    # for pair in pairs:
    #     print(f"Rows {pair[0]} and {pair[1]} differ in columns {pair[2]}")
    
    return pairs
```

```python
def find_row_pairs_BikeSP(df, num_diff):
    pairs = []

    # iterate over each pair of rows in the dataframe
    for i, row1 in df.iterrows():
        for j, row2 in df.iterrows():
            if i < j:  # avoid checking same row pair twice
                diff = (row1 != row2)  # boolean array of differences between rows
                if diff.sum() == num_diff:
                    cols = diff[diff].index.tolist()  # get column names where the rows differ
                    if 'BikeSP' in cols:
                        pairs.append((i, j, cols))
                        for col in cols:
                            if col == 'BikeSP':
                                print("The difference in the column is " + str(row1[col]) +' vs ' + str(row2[col]))
    # for pair in pairs:
    #     print(f"Rows {pair[0]} and {pair[1]} differ in columns {pair[2]}")
    
    return pairs
```
