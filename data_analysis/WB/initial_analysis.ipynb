```python

def define_scenario_columns(scenario_number):
    scenario_columns_ficha1 = [f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Tipo de ficha)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Selecionado)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Tempo de viagem (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Tempo de Acesso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Tempo de Egresso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Custo (R$))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Presenca de infraestrutura cicloviária)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Remuneracao Bike SP)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Estacionamento de bicicleta)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 1 - Facilidades no destino)'.format(scenario_number = scenario_number),
    ]
    scenario_columns_ficha2 = [
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Tipo de ficha)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Selecionado)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Tempo de viagem (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Tempo de Acesso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Tempo de Egresso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Custo (R$))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Presenca de infraestrutura cicloviária)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Remuneracao Bike SP)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Estacionamento de bicicleta)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 2 - Facilidades no destino)'.format(scenario_number = scenario_number),
    ]
    scenario_columns_ficha3 = [
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Tipo de ficha)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Selecionado)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Tempo de viagem (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Tempo de Acesso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Tempo de Egresso (min))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Custo (R$))'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Presenca de infraestrutura cicloviária)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Remuneracao Bike SP)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Estacionamento de bicicleta)'.format(scenario_number = scenario_number),
        f'Ciclista - PD Cenário {scenario_number} (Ficha 3 - Facilidades no destino)'.format(scenario_number = scenario_number),
    ]
    return scenario_columns_ficha1, scenario_columns_ficha2, scenario_columns_ficha3
```
