```python
# Developing a function to round to a multiple
def round_to_multiple_5(number):
    multiple = 5
    return multiple * round(number / multiple)
```

#### Creating the missing bike columns

```python
bike = bike.rename({'distancia_m': 'distancia (metros)'}, axis = 1)
bike = bike.rename({'tempo_m': 'tempo (min)'}, axis = 1)
bike['tempo_aprox (min)'] = bike['tempo (min)'].apply(round_to_multiple_5)
bike['distancia_aprox (km)'] = round(bike['distancia (metros)'] / 1000, 1)
```
