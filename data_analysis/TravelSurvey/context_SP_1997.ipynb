## Mode share

Columns correspondence:

01 - Ônibus

02 - Ônibus Fretado

03 - Transporte Escolar

04 - Dirigindo Automóvel

05 - Passageiro de Automóvel

06 - Táxi

07 - Lotação/Perua

08 - Metrô

09 - Trem

10 - Moto

11 - Bicicleta

12 - A Pé

13 - Outros


```python
extended_mode_column = 'MODOPRIN'
extended_mode_index_map = {1: '01 - Ônibus', 2: '02 - Ônibus Fretado', 3: '03 - Transporte Escolar',
                  4:'04 - Dirigindo Automóvel', 5: '05 - Passageiro de Automóvel',
                  6: '06 - Táxi', 7: '07 - Lotação/Perua', 8: '08 - Metrô',
                  9: '09 - Trem', 10: '10 - Moto', 11: '11 - Bicicleta',
                  12:'12 - A Pé', 13: '13 - Outros'}
df_modes = tsa.calculate_distribution(data, extended_mode_column, expansion_factor_trip, extended_mode_index_map)
df_modes
```

```python
collective_filter = ['01 - Ônibus', '02 - Ônibus Fretado', '03 - Transporte Escolar',
                    '07 - Lotação/Perua', '08 - Metrô', '09 - Trem']

individual_filter = ['04 - Dirigindo Automóvel', '05 - Passageiro de Automóvel',
                    '06 - Táxi', '10 - Moto', '13 - Outros']
walk_filter = ['12 - A Pé']
bike_filter = ['11 - Bicicleta']
```

```python
mode_column = 'TIPO_VG'
mode_index_map = {1: 'Collective', 2: 'Individual', 3: 'On foot'}
tsa.calculate_distribution(data, mode_column, expansion_factor_trip, mode_index_map)
```

```python
df_modes.loc[individual_filter].sum() + df_modes.loc[bike_filter].sum()
```

```python
df_modes.loc[walk_filter].sum() 
```

```python
personalized_mode_index_map = {}

for item in collective_filter:
    personalized_mode_index_map[item] = 'Collective'
for item in individual_filter:
    personalized_mode_index_map[item] = 'Individual'
for item in walk_filter:
    personalized_mode_index_map[item] = 'On foot'
for item in bike_filter:
    personalized_mode_index_map[item] = 'Bicycle'

personalized_mode_index_map
```

```python
convert_to_new_mode_index_map = {1: 1, 2: 5, 3: 5, 4: 2, 5: 2, 6: 2, 7: 5, 8: 1, 9: 1, 10: 5, 11: 3, 12: 4, 13: 5}
new_mode_index_map = {1: 'Public Transportation', 2: 'Car', 3: 'Bicycle', 4: 'On foot', 5: 'Other'}
```

```python
new_mode_column = extended_mode_column + '_reduced'
data_sp[new_mode_column] = data_sp[extended_mode_column].map(convert_to_new_mode_index_map)
```
