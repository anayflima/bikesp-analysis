## Mode share

Columns correspondence:

03 - Transporte Escolar

04 - Dirigindo Automóvel

05 - Passageiro de Automóvel

06 - Táxi

07 - Lotação/Perua

08 - Metrô

09 - Trem

10 - Moto

11 - Bicicleta

12 - A Pé

13 - Outros


```python
mode_column = 'MODOPRIN'
mode_index_map = {1: '1 - Ônibus Trólebus', 2: '2 - Ônibus Escolar/Empresa', 3: '3 - Dirigindo Automóvel',
                  4:'4 - Passageiro de Automóvel', 5: '5 - Táxi',
                  6: '6 - Lotação/Perua', 7: '7 - Metrô', 8: '8 - Trem',
                  9: '9 - Motocicleta', 10: '10 - Bicicleta', 11: '11 - A Pé',
                  12: '12 - Outros'}
df_modes = tsa.calculate_distribution(data, mode_column, expansion_factor_trip, mode_index_map)
df_modes
```

```python
collective_filter = ['1 - Ônibus Trólebus', '2 - Ônibus Escolar/Empresa',
                    '6 - Lotação/Perua', '7 - Metrô', '8 - Trem']

individual_filter = ['3 - Dirigindo Automóvel', '4 - Passageiro de Automóvel', '5 - Táxi',
                    '9 - Motocicleta', '12 - Outros']
walk_filter = ['11 - A Pé']
bike_filter = ['10 - Bicicleta']
```

```python
mode_column = 'TIPO_VG'
type_index_map = {1: 'Collective', 2: 'Individual', 3: 'On foot'}
tsa.calculate_distribution(data, mode_column, expansion_factor_trip, type_index_map)
```

```python
df_modes.loc[walk_filter].sum() 
```

```python
mode_index_map = {}

for item in collective_filter:
    mode_index_map[item] = 'Collective'
for item in individual_filter:
    mode_index_map[item] = 'Individual'
for item in walk_filter:
    mode_index_map[item] = 'On foot'
for item in bike_filter:
        mode_index_map[item] = 'Bicycle'

mode_index_map
```
