```python
mode_columns = ['MODO1', 'MODO2', 'MODO3', 'MODO4', 'MODOPRIN']
```

```python
bike_trips_multimodal[mode_columns]
```

```python
extended_mode_index_map = {1: '01 - Metrô', 2: '02 - Trem', 3:'03 - Monotrilho',4:'04 - Ônibus/micro-ônibus/perua do município de São Paulo',
                  5:'05 - Ônibus/micro-ônibus/perua de outros municípios',6: '06 - Ônibus/micro-ônibus/perua metropolitano',
                  7:'07 - Transporte Fretado',8:'08 - Transporte Escolar',9:'09 - Dirigindo Automóvel',10:'10 - Passageiro de Automóvel',
                  11: '11 - Táxi Convencional',12:'12 - Táxi não Convencional', 13: '13 - Dirigindo Moto', 14: '14 - Passageiro de Moto',
                  15: '15 - Bicicleta', 16: '16 - A Pé', 17: '17 - Outros'}
```

```python
for column in mode_columns:
    bike_trips_multimodal[column] = bike_trips_multimodal[column].map(extended_mode_index_map)
```

```python
bike_trips_multimodal[mode_columns]
```

```python
bike_trips_multimodal[mode_columns] = bike_trips_multimodal[mode_columns].fillna('-')
```

```python
bike_trips_multimodal.groupby(['MODO1', 'MODO2', 'MODO3', 'MODO4']).size().reset_index(name='Number of occurrences').sort_values(by = 'Number of occurrences', ascending = False)
```

```python
#define a custom function that sorts the values in each row
def sort_row(row):
    return tuple(sorted(row))

df = bike_trips_multimodal

#apply the function to the four mode columns and create a new column called 'sorted_modes'
df['sorted_modes'] = df[['MODO1', 'MODO2', 'MODO3', 'MODO4']].apply(sort_row, axis=1)

#group by the new column and count the size of each group
df_count = df.groupby('sorted_modes').size().reset_index(name='Number of occurrences')
df_count
```
