## Duration of walk trips in different social classes

```python
def calculate_distribution_per_social_class(data, social_class_column, variable_column, index_map):
    df = pd.DataFrame()
    for social_class in index_map.keys():
        data_class = data[data[social_class_column] == social_class]
        describe = data_class[variable_column].describe()
        df = df.append(describe, ignore_index=True)
    df.index = index_map.values()
    return df
```
